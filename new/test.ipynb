```python
from PIL import Image

def center_crop(image, new_width, new_height):
    width, height = image.size
    left = (width - new_width) // 2
    top = (height - new_height) // 2
    right = (width + new_width) // 2
    bottom = (height + new_height) // 2
    return image.crop((left, top, right, bottom))

def bicubic_downsample(image, scale_factor):
    width, height = image.size
    new_width = int(width / scale_factor)
    new_height = int(height / scale_factor)
    resized_image = image.resize((new_width, new_height), Image.BICUBIC)
    return resized_image

# Load the image
input_image_path = r"C:\Users\Marcelio Darel Liem\thesis\Video-Enhancement-With-GAN\Dataset\vimeo-septuplet\sequences\00026\0018\im5.png"
output_image_path = "output_image.jpg"
input_image = Image.open(input_image_path)

# Downsample by a factor of 4 using bicubic interpolation
downsampled_image = bicubic_downsample(input_image, 4)

# Center crop
new_width, new_height = 64, 64
center_cropped_image = center_crop(downsampled_image, new_width, new_height)

# Save the downsampled image
center_cropped_image.save(output_image_path)
```
